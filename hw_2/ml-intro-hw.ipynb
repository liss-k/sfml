#### Задание 3. Фунция предобработки

Напишите функцию, которая бы

* Удаляло идентификатор `AGREEMENT_RK`
* Избавлялась от проблем с '.' и ',' в стобцах PERSONAL_INCOME, CREDIT, FST_PAYMENT, LOAN_AVG_DLQ_AMT, LOAN_MAX_DLQ_AMT
* Что-то делала с пропусками
* Кодировала категориальные признаки

В результате, ваш датафрейм должен содержать только числа и не содержать пропусков!

```python
def family_income (s):
    if s == 'до 5000 руб.':
        return 0
    elif s == 'от 5000 до 10000 руб.':
        return 1
    elif s == 'от 10000 до 20000 руб.':
        return 2
    elif s == 'от 20000 до 50000 руб.':
        return 3
    elif s == 'свыше 50000 руб.':
        return 4


def preproc_data(df_input):
    df_output = df_input.copy()
    
    df_output = df_output.drop(['AGREEMENT_RK','REG_ADDRESS_PROVINCE','POSTAL_ADDRESS_PROVINCE'], axis=1)
## ----------- test ---------
#     df_output = df_output.drop(['GEN_INDUSTRY','GEN_TITLE','ORG_TP_STATE','ORG_TP_FCAPITAL','JOB_DIR', 'TP_PROVINCE'], axis=1)
## --------------------------    
    col_type_to_float=['PERSONAL_INCOME', 'CREDIT', 'FST_PAYMENT', 'LOAN_AVG_DLQ_AMT', 'LOAN_MAX_DLQ_AMT']
    for i in col_type_to_float:
        df_output[i]=df_output[i].map(lambda x: x.replace(',', '.')).astype('float')
    
    
    df_output['FAMILY_INCOME']=df_output['FAMILY_INCOME'].map(lambda s: family_income(s))
    df_output.loc[df_output['PREVIOUS_CARD_NUM_UTILIZED'].isnull(),'PREVIOUS_CARD_NUM_UTILIZED']=0
    df_output.loc[df_output['WORK_TIME'].isnull(),'WORK_TIME']=0 

    to_kill_null=['GEN_INDUSTRY','GEN_TITLE','ORG_TP_STATE','ORG_TP_FCAPITAL','JOB_DIR', 'TP_PROVINCE']
    for i in to_kill_null:
        df_output.loc[df_output[i].isnull(),i]='не указано'
        
    df_output=pd.get_dummies(df_output)
    
    
    return df_output
```

```python
df_preproc = df.pipe(preproc_data)
```

#### Задание 5. Обучение и оценка качества разных моделей

```python
from sklearn.model_selection import train_test_split
# test_size=0.3, random_state=42

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Попробовать следующие "черные ящики": интерфейс одинаковый 
#     fit, 
#     predict, 
#     predict_proba

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

dtc = DecisionTreeClassifier(random_state=42)
dtc.fit(X_train, y_train)
predict1 = dtc.predict(X_test)

rfc = RandomForestClassifier(random_state=42)
rfc.fit(X_train, y_train)
predict2 = rfc.predict(X_test)

lr = LogisticRegression(random_state=42)
lr.fit(X_train, y_train)
predict3 = lr.predict(X_test)

# from sklearn.metrics import confusion_matrix
# confusion_matrix = confusion_matrix(y_test, predict1)
# print(confusion_matrix)
```

```python
## Посчитать метрики стандартные
# accuracy, precision, recall

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print('precision1 ', precision_score(y_test, predict1))
print('recall1 ', recall_score(y_test, predict1))
print('accuracy1 ', accuracy_score(y_test, predict1))
print('f1_1 ', f1_score(y_test, predict1))  
print('\n')
print('precision2 ', precision_score(y_test, predict2))
print('recall2 ', recall_score(y_test, predict2))
print('accuracy2 ', accuracy_score(y_test, predict2))
print('f1_2 ', f1_score(y_test, predict2)) 
print('\n')
print('precision3 ', precision_score(y_test, predict3))
print('recall3 ', recall_score(y_test, predict3))
print('accuracy3 ', accuracy_score(y_test, predict3))
print('f1_3', f1_score(y_test, predict3)) 
```

```python
rfc = RandomForestClassifier(random_state=42)
rfc.fit(X_train, y_train)
predict22 = rfc.predict_proba(X_test)
```

```python
# Визуалищировать эти метрики всех моделей на одном графике (чтоб визуально посмотреть)

from sklearn.metrics import precision_recall_curve
precision1, recall1, treshholds1 = precision_recall_curve(y_test,predict1)
precision22, recall22, treshholds22 = precision_recall_curve(y_test,predict22[:,1])
precision2, recall2, treshholds2 = precision_recall_curve(y_test,predict2)
precision3, recall3, treshholds3 = precision_recall_curve(y_test,predict3)

from matplotlib import pyplot as plt
plt.figure(figsize=(5, 5))
plt.plot(precision1, recall1)
plt.plot(precision22, recall22)
plt.plot(precision2, recall2)
plt.plot(precision3, recall3)
plt.ylabel('recall')
plt.xlabel('precision')
plt.grid(True)
plt.title('PR curve')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
```

```python
# Потроить roc-кривые всех можелей на одном графике
# Вывести roc_auc каждой моделе

from matplotlib import pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve

fpr1, tpr1, thresholds1 = roc_curve(y_test, predict1)
fpr22, tpr22, thresholds22 = roc_curve(y_test, predict22[:,1])
fpr2, tpr2, thresholds2 = roc_curve(y_test, predict2)
fpr3, tpr3, thresholds3 = roc_curve(y_test, predict3)
plt.figure(figsize=(5, 5))
plt.plot(fpr1, tpr1, label='dtc')
plt.plot(fpr22, tpr22, label='rfc_proba')
plt.plot(fpr2, tpr2, label='rfc')
plt.plot(fpr3, tpr3, label='logreg')
plt.legend(loc='upper left')
plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
```
