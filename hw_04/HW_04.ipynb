```python
import pymorphy2
morph=pymorphy2.MorphAnalyzer()
def getMeaningfullWords(text):
    words=[]
    tokens=re.findall('[А-Яа-яЁё]+\-[А-Яа-яЁё]+|[А-Яа-яЁё]+', text)
    for t in tokens:
        pv=morph.parse(t)  
        for p in pv:
            if p.tag.POS in ['ADJF', 'NOUN', 'VERB']:
                words.append(p.normal_form)
                break
    return words

def get_list(text):
    c=[' '.join(getMeaningfullWords(text))]
    return c
```
